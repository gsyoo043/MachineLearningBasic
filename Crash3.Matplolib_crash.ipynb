## multiple subplots

**y = x + noise 인 (x, y) 쌍의 toy data 생성**

````python
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)

data = np.column_stack((x, y))
data[:5]
````

````python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

ax1.scatter(x, y, marker='o', c='r', edgecolor='b')
ax1.set_title("Scatter: x vs. y")
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.hist(data, bins=10, label=('x', 'y'))
ax2.set_title("Frequency")
ax2.legend()
ax2.yaxis.tick_right()
````

## meshgrid / contour (3 차원 자료의 2 차원 표현)

- 입력 변수가 x, y 두 개이고 출력 변수가 z 하나인 경우에는 3차원 자료가 됨   


- 3차원 자료를 2차원 평면에 시각화하는 방법 


    - 색 (color) 로 구분  
    - 등고선(contour) 으로 표시 -  contour 혹은 contourf 명령 사용 (contour는 등고선만 표시하고 contourf는 색깔로 표시)
    - 입력 변수 x, y는 그대로 사용할 수 없고 meshgrid 명령으로 그리드 포인트 행렬을 만들어야 함
    

- meshgrid : coordinate vector 를 이용하여 coordinate matrix 구성

<img src="meshgrid.png" width="600">


````python
x = np.array([1, 2, 3, 4])
y = np.array([7, 6, 5])

XX, YY = np.meshgrid(x, y)
print(XX)
print()
print(YY)
````

<img src="meshgrid1.png" width="250">

````python
x = np.array([0, 1, 2])
y = np.array([0, 1, 2])
XX, YY = np.meshgrid(x, y)

for i in range(len(XX)):
    for x, y in zip(XX[i], YY[i]):
        print((x, y))
    print()
````
